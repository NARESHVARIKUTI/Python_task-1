```python
customer_data = pd.read_csv(r"C:\Python\us_customer_data 1 1.csv")
```

```python
customer_data = customer_data.drop_duplicates(subset="customer_id", keep="first")
```

```python
customer_data.isna().sum()
```

```python
customer_data["name"] = customer_data["name"].astype(str).str.strip().str.title()
```

```python
customer_data["name"]
```

```python
customer_data["email"].isna().sum()
```

```python
customer_data["email"] = customer_data.apply(
    lambda row: f"{row["name"].lower().replace(" ", "")}@gmail.com"
    if pd.isna(row["email"]) else (row["email"]),
axis=1
)
```

```python
customer_data["email"].isna().sum()
```

```python
customer_data.head(28)
```

```python
customer_data["phone"].isna().sum()
```

```python
customer_data["phone"] = customer_data["phone"].str.replace(r"[A-Za-z]", '', regex="true")
```

```python
customer_data["phone"] = customer_data.apply(
    lambda row: "000-000-0000" if pd.isna(row["phone"]) else row["phone"],
    axis=1
)
```

```python
customer_data.head(25)
```

```python
customer_data["phone"] = customer_data["phone"].str.replace(r"\D", "", regex=True)
```

```python
customer_data["phone"] = customer_data.apply(
    lambda row: f"{row["phone"][:3]} ({row["phone"][3:6]})-{row["phone"][6:10]}"
    if len(row["phone"]) >= 10 else'null',
    axis = 1
    )
```
