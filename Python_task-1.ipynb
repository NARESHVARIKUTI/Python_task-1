```python
df_unique = df.drop_duplicates(subset="customer_id", keep="first") #Removing Duplicates Using Customer ID
```

```python
df["name"] = df["name"].astype(str).str.strip().str.title() #Clean Customer Names, Remove extra spaces and Convert to Proper Case
```

```python
df["email"].isna().sum() # it filters null values in the email (specific) column 
```

```python
df["email"] = df["email"].astype(str).str.strip().str.title() #Clean Customer Names, Remove extra spaces and Convert to Proper Case
```

```python
df["email"] = df.apply(
    lambda row: f"{row['name'].lower().replace(' ', '')}@gmail.com"
    if pd.isna(row["email"]) else row["email"].lower(),
    axis=1
)
```

```python
df["phone"] = df["phone"].astype(str).str.strip().str.title() ##Clean email, Remove extra spaces and Convert to Proper Case
```

```python
df["phone"] = df["phone"].str.replace(r"[A-Za-z]", "", regex="true") ## removing letteres in the rows in phone column
```

```python
df["phone"] = df.apply (
    lambda row:"000-000-0000" if pd.isnull(row["phone"]) else(row["phone"]), axis=1) # adding phone numbers in place of null values
```
