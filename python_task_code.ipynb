```python
df = pd.read_csv(r"C:\Python\us_customer_data 1 1.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_unique = df.drop_duplicates(subset="customer_id", keep="first") #Removing Duplicates Using Customer ID
```

```python
df.head()
```

```python
df["name"] = df["name"].astype(str).str.strip().str.title() #Clean Customer Names, Remove extra spaces and Convert to Proper Case
```

```python
df.head()
```

```python
df["email"] = df["name"].astype(str).str.strip().str.title() #Clean email, Remove extra spaces and Convert to Proper Case
```

```python
df.head()
```

```python
df["email"] = df.apply(
    lambda row: f"{row["name"]}@gmail.com".lower()
    if pd.isnull(row["email"]) else(row["email"]), 
axis=1
)
```

```python
df.isna().sum()
```

```python
df["phone"] = df["phone"].astype(str).str.strip().str.title() ##Clean email, Remove extra spaces and Convert to Proper Case
```

```python
df.info()
```

```python
df["phone"] = df["phone"].str.replace(r"[A-Za-z]", "", regex="true") ## removing letteres in the rows in phone column
```

```python
df["phone"] = df.apply (
    lambda row:"000-000-0000" if pd.isnull(row["phone"]) else(row["phone"]), axis=1) # adding phone numbers in place of null values
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df["phone"].notnull()]
df = df[df["customer_id"].notnull()] #Removing Invalid Rows
```

```python
df.to_csv(r"C:\Python\customer_data_cleaned.csv", index=False)
```
